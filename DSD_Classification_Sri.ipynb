```python
features = df[['ConversionFactor','ConversionResidual','CategoryID','InventoryChange','ActualScans','ForecastedScans','WeightData','BaseOrder','BaseorderID','SlowMoving','TrueUpAdjQty','TooBig','ConversionUnits','targetsize','Original','TUAIssue','OriginalPropOrderQty','target']]
Labels = df[['targetsize']]
```

```python
train_dataset = features.sample(frac=0.8, random_state=0)
test_dataset = (Labels.index)
```

```python
def input_fn(features, Labels, training=True, batch_size=256):
    # Convert the inputs to a Dataset.
    dataset = tf.data.Dataset.from_tensor_slices((dict(features), Labels))

    # Shuffle and repeat if you are in training mode.
    if training:
        dataset = dataset.shuffle(1000).repeat()
    
    return dataset.batch(batch_size)
```

```python
# Feature columns describe how to use the input.
my_feature_columns = []
for key in features.keys():
    my_feature_columns.append(tf.feature_column.numeric_column(key=key))
print(my_feature_columns)
```

```python
# Build a DNN with 2 hidden layers with 30 and 10 hidden nodes each.
classifier = tf.estimator.DNNClassifier(
    feature_columns=my_feature_columns,
    # Two hidden layers of 30 and 10 nodes respectively.
    hidden_units=[3, 10,0],
    # The model must choose between 3 classes.
    n_classes=3)
```

```python
classifier.train(
    input_fn=lambda: input_fn(features,Labels,training=True),
    steps=5000)
```

```python
eval_result = classifier.evaluate(
    input_fn=lambda: input_fn(features,Labels, training=False))

print('\nTest set accuracy: {accuracy:0.3f}\n'.format(**eval_result))
```
